```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
x=np.array([1920,1930,1940,1950,1960,1970,1980,1990]) #Years
y=np.array([106.46,123.08,132.12,152.27,180.67,205.05,227.23,249.46]) #Consumption of Water liters
```

```python
year=2000
coef=np.polyfit(x,y,2)
p=np.polyval(coef,year)
print(f"For the year {year} the prediction is {p.round(2)}")
```

```python
def fx (x1, coef):
    fx = 0
    n = len(coef) - 1
    for p in coef:
        fx = fx + p*x1**n
        n = n - 1
    return fx
```

```python
year = 2000
for i in range(0,10):
    coef = np.polyfit(x,y,i)
    p = np.polyval(coef, year)
    
    print(f"For grade {i} the prediction is: {p.round(2)}")
    x1 = np.linspace(1920, year + 1, 1000)
    y1 = fx(x1, coef) # funcion
    plt.figure(figsize=[20,10])
    plt.title(f"Quantity of liters vs Year:{str(i)} ")
    
    plt.scatter(x,y,s=120,c='blueviolet')
    plt.plot(x1,y1,"--",linewidth=3,color='orange')
    plt.scatter(year,p,s=200,c='red')
    plt.yticks(range(100,320,20))
    plt.grid("on")
    ax=plt.gca()
    ax.set_xlabel("$Years$")
    ax.set_ylabel("$Quantity Liters$")
    #plt.savefig("img" + str(i)+".jpg", dpi=600)
    plt.show()
```

What grade is better to use? to answer that question we have to use mean square error MSE

```python
year = 2000
grade = np.arange(0, 30 + 1, 1)
aproxi = np.array([])
y_pred_vec = np.array([])
for i in grade:
    coef = np.polyfit(x,y,i)
    p = np.polyval(coef, year)
    aproxi = np.append(aproxi, p)
    
    # For MSE
    y_pred_vec = np.array([])
    for j in x:
        y_pred = np.polyval(coef, j)
        y_pred_vec = np.append(y_pred_vec,y_pred)
    #print(f"los y: {y}")
    #print(f"los y_aprox: {y_pred_vec}")
    MSE = (sum( (y - y_pred_vec)**2 ))/len(y)
    print(f"For grade {i} The MSE is: {MSE}")
plt.figure(figsize=[20,10])
plt.title("MSE vs Prediction")
plt.plot(grade,aproxi,"--",linewidth=3,color='red')
plt.grid("on")
plt.show()
```
