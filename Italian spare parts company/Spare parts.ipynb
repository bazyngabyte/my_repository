### 2. Cleaning up the Data!

#### 2.1 Cleaning Sales Data

```python
df.head(10)
```

```python
# Deleting column 3

del df[3]
```

```python
df.head(10)
```

```python
df.columns
```

```python
# Renaiming columns

df=df.rename(columns=({0:"Order",1:"date",2:"seller",4:"cat-prod",5:"price",6:"quantity",7:"region"}))
df.sample()
```

```python
# Dropping useless rows

df=df.loc[6:]
df
```

```python
# Checking nulls

df.isnull().sum()
```

```python
filter=df.seller.isnull()
df.loc[filter]
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
# Checking the type of columns

df.info()
```

```python
# Casting columns

df["date"]=pd.to_datetime(df["date"]).dt.date
df
```

```python
df["price"]=pd.to_numeric(df["price"]).astype(float)
```

```python
df["quantity"]=pd.to_numeric(df["quantity"]).astype(int)
```

```python
df.seller.unique()
```

```python
df["seller"]=df.seller.str.strip()
df["seller"]=df.seller.str.title()
```

```python
df.groupby("region").quantity.sum()
```

```python
# Dropping "-" in Trentino-Alto-Adige

df["region"]=df["region"].str.replace("-"," ")
```

```python
df.info()
```

```python
# Adding revenue,tax and total columns

df.insert(6,"revenue","")
```

```python
df.insert(7,"tax","")
```

```python
df.insert(8,"total","")
```

```python
df
```

```python
# Calculating columns

df=df.assign(revenue=lambda x:(x["price"]*x["quantity"]))
```

```python
df=df.assign(tax=lambda x:(x["revenue"]*0.22))
```

```python
df=df.assign(total=lambda x:(x["revenue"]+x["tax"]))
```

```python
df
```

```python
df.insert(4,"product","")
```

```python
df
```

```python
# Splitting Category and product

def get_product(product):
    return product.split("-")[1]
def get_category(product):
    return product.split("-")[0]

df["product"]=df["cat-prod"].apply(lambda x:get_product(x))
df["cat-prod"]=df["cat-prod"].apply(lambda x:get_category(x))
```

```python
df
```

```python
df=df.rename(columns=({"cat-prod":"category"}))
df.sample()
```

```python
# Checking for zeros

filter=df.total==0
df.loc[filter]
```

```python
# Dropping Zeros

filter=df.total!=0
df=df.loc[filter]
df
```

```python
df["product"]=df["product"].str.capitalize()
df
```

### 3. Analyse Data

```python
# Making copies

sales=df.copy()
```

```python
sales
```

#### 3.1 Sales by sellers

```python
seller=sales.groupby(["seller"],as_index=False).revenue.sum()
seller
```

```python
seller["revenue"]=seller.revenue.round(2)
seller
```

```python
df_seller
```

```python
sales_seller=pd.merge(left=seller,right=df_seller,on="seller")
sales_seller
```

```python
sales_seller.insert(3,"delta","")
sales_seller
```

```python
sales_seller=sales_seller.assign(delta=lambda x:((x["revenue"]-x["target"])/x["target"])).round(2).sort_values("revenue",ascending=False)
sales_seller
```

#### 3.2 Sales by Region

```python
region_sales=sales.groupby(["region"],as_index=False).revenue.sum()
region_sales
```

```python
df_province
```

```python
sales_region=pd.merge(left=region_sales,right=df_province,on="region")
sales_region
```
