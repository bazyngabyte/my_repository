### 2. Cleaning up the Data!

#### 2.1. Cleaning Speedband Table

```python
df
```

```python
df.rename(columns=str.lower,inplace=True)
df.sample()
```

```python
df.columns=df.columns.str.replace(" ","_")
df.sample()
```

```python
df.columns
```

```python
filtercolumns=["country","countrycode","mean_download_speed_(mbps)","region"]
df=df.loc[:,filtercolumns]
df
```

```python
df.rename(columns=({"mean_download_speed_(mbps)":"mean_speed"}),inplace=True)
df.sample()
```

```python
df
```

```python
filter=df.countrycode.isnull()
df.loc[filter]
```

```python
df["countrycode"]=df.countrycode.fillna("NA")
df
```

```python
df["mean_speed"]=df.mean_speed.round(2)
df
```

```python
df.region.unique()
```

#### 2.2 Cleaning Broadband Cost table

```python
dfp
```

```python
dfp=dfp[1:]
dfp.head(10)
```

```python
dfp.iloc[0]
```

```python
filtercolumns=[2,8,14]
dfp=dfp.loc[:,filtercolumns]
dfp.head(5)
```

```python
dfp.rename(columns=({2:"country",8:"cost_broadband_(USD)",14:"cost_megabit_(USD)"}),inplace=True)
dfp.head(10)
```

```python
dfp=dfp.iloc[1:]
dfp.head(10)
```

```python
dfp["cost_broadband_(USD)"]=pd.to_numeric(dfp["cost_broadband_(USD)"])
dfp["cost_megabit_(USD)"]=pd.to_numeric(dfp["cost_megabit_(USD)"])
dfp
```

```python
dfp["cost_broadband_(USD)"]=dfp["cost_broadband_(USD)"].round(2)
dfp["cost_megabit_(USD)"]=dfp["cost_megabit_(USD)"].round(2)
dfp
```

```python
df.sample(10)
```

#### 2.3 Merging tables

```python
dfm=pd.merge(left=df,right=dfp,on="country")
dfm
```

```python
dfm=dfm.reindex(columns=(["region","countrycode","country","mean_speed","cost_broadband_(USD)","cost_megabit_(USD)"]))
dfm
```

```python
dfm.sample()
```

```python
dfm=dfm.rename(columns=({"cost_broadband_(USD)":"broadband_cost","cost_megabit_(USD)":"megabit_cost"}))
dfm
```

### 3. Analyse Data

```python
# Making copies

internet=dfm.copy()
```

```python
internet
```

### 3.1 Top 10: Fastest internet by country

```python
fastest=internet.groupby(["country"],as_index=False).mean_speed.sum().sort_values("mean_speed",ascending=False)
fastest
```

```python
plt.figure(figsize=(10,6))
ax=sns.barplot(x="country",y="mean_speed",data=fastest[:10],hue="country",palette="tab10")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("(2023) Top 10: The Fastest Internet by Country.")
plt.xlabel("Country",fontsize=12)
plt.ylabel("Mean Speed (Mbps)",fontsize=12)
plt.xticks(rotation=90,size=9)
plt.show()
```

### 3.2 Top 10: Slowest internet by country

```python
slowest=internet.groupby(["country"],as_index=False).mean_speed.sum().sort_values("mean_speed",ascending=True)
slowest
```

```python
plt.figure(figsize=(10,6))
ax=sns.barplot(x="country",y="mean_speed",data=slowest[:10],hue="country",palette="tab10")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("(2023) Top 10: The Slowest Internet by Country")
plt.xlabel("Country",fontsize=12)
plt.ylabel("Mean Speed (Mbps)",fontsize=12)
plt.xticks(rotation=90,size=9)
plt.show()
```

```python
region=internet.groupby(["region"],as_index=False).mean_speed.mean().sort_values("mean_speed",ascending=False)
region
```

```python
plt.figure(figsize=(10,6))
ax=sns.barplot(x="mean_speed",y="region",data=region[:10],hue="region",palette="tab10")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("(2023) The internet speed by Region")
plt.xlabel("Mean Speed (Mbps)",fontsize=12)
plt.ylabel("Country",fontsize=12)
plt.xticks(rotation=90,size=9)
plt.show()
```

### 3.3 Top 10: The most expensive internet by Country

```python
internet
```

```python
expensive=internet.groupby(["country"],as_index=False).broadband_cost.sum().sort_values("broadband_cost",ascending=False)
expensive
```

```python
plt.figure(figsize=(10,6))
ax=sns.barplot(x="country",y="broadband_cost",data=expensive[:10],hue="country",palette="tab10")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("(2023) Top 10: The most expensive Internet by Country")
plt.xlabel("Country",fontsize=12)
plt.ylabel("Broadband Cost (USD)",fontsize=12)
plt.xticks(rotation=90,size=9)
plt.show()
```

### 3.4 Top 10: The cheapest internet by Country

```python
cheapest=internet.groupby(["country"],as_index=False).broadband_cost.sum().sort_values("broadband_cost",ascending=True)
cheapest
```

```python
plt.figure(figsize=(10,6))
ax=sns.barplot(x="country",y="broadband_cost",data=cheapest[:10],hue="country",palette="tab10")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("(2023) Top 10: The cheapest Internet by Country")
plt.xlabel("Country",fontsize=12)
plt.ylabel("Broadband Cost (USD)",fontsize=12)
plt.xticks(rotation=90,size=9)
plt.show()
```

### 3.5 Average Broadband Cost by Region

```python
cost_region=internet.groupby(["region"],as_index=False).broadband_cost.mean().round(2).sort_values("broadband_cost",ascending=False)
cost_region
```

```python
plt.figure(figsize=(10,6))
ax=sns.barplot(x="broadband_cost",y="region",data=cost_region[:10],hue="region",palette="tab10")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("(2023) The Broadband Cost by Region")
plt.xlabel("Broadband Cost(USD) average",fontsize=12)
plt.ylabel("Region",fontsize=12)
plt.xticks(rotation=90,size=9)
plt.show()
```
